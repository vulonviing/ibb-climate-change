```python
#önce koordinatları çekiyorum


from geopy.geocoders import Nominatim

# İl isimleri
il_names = ["Adana", "Adıyaman", "Afyonkarahisar", "Ağrı", "Aksaray", "Amasya", "Ankara", "Antalya", "Ardahan", "Artvin", "Aydın", "Balıkesir", "Bartın", "Batman", "Bayburt", "Bilecik", "Bingöl", "Bitlis", "Bolu", "Burdur", "Bursa", "Çanakkale", "Çankırı", "Çorum", "Denizli", "Diyarbakır", "Düzce", "Edirne", "Elazığ", "Erzincan", "Erzurum", "Eskişehir", "Gaziantep", "Giresun", "Gümüşhane", "Hakkâri", "Hatay", "Iğdır", "Isparta", "İstanbul", "İzmir", "Kahramanmaraş", "Karabük", "Karaman", "Kars", "Kastamonu", "Kayseri", "Kırıkkale", "Kırklareli", "Kırşehir", "Kilis", "Kocaeli", "Konya", "Kütahya", "Malatya", "Manisa", "Mardin", "Mersin", "Muğla", "Muş", "Nevşehir", "Niğde", "Ordu", "Osmaniye", "Rize", "Sakarya", "Samsun", "Siirt", "Sinop", "Sivas", "Şanlıurfa", "Şırnak", "Tekirdağ", "Tokat", "Trabzon", "Tunceli", "Uşak", "Van", "Yalova", "Yozgat", "Zonguldak"]

geolocator = Nominatim(user_agent="location-finder")

# İl isimleri ve koordinatları
il_coordinates = {}

for il in il_names:
    location = geolocator.geocode(il + ", Türkiye")
    if location:
        il_coordinates[il] = (location.latitude, location.longitude)
    else:
        il_coordinates[il] = None

print(il_coordinates)
```

```python
#koordinatlara göre türkiye hava durumunu çekiyorum

# Import Meteostat library and dependencies
from datetime import datetime
import matplotlib.pyplot as plt
from meteostat import Point, Daily, Hourly, Monthly, Normals
import pandas as pd

türkiye_daily = []
türkiye_monthly = []

for city, coordinates in il_coordinates.items():
    # Set time period
    start = datetime(1929, 1, 1)
    end = datetime(2023, 8, 25)

    #set location
    location = Point(coordinates[0], coordinates[1])

    # Get daily data 
    data_daily = Daily(location, start, end)
    data_daily = data_daily.fetch()
    data_daily["city"] = city
    data_daily.reset_index(inplace=True)

    # Get monthly data 
    data_monthly = Monthly(location, start, end)
    data_monthly = data_monthly.fetch()
    data_monthly["city"] = city
    data_monthly.reset_index(inplace=True)
    
    türkiye_daily.append(data_daily)
    türkiye_monthly.append(data_monthly)

# Concatenate all dataframes into a single dataframe
türkiye_daily = pd.concat(türkiye_daily, ignore_index=False)
türkiye_monthly = pd.concat(türkiye_monthly, ignore_index=False)
türkiye_daily["country"] = "Türkiye"
türkiye_monthly["country"] = "Türkiye"
```
